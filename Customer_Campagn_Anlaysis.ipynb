```python
data['EmailOpenRate']=np.where(data['EmailOpens']>0, data['EmailClicks'] / data['EmailOpens'],
    0)
```

```python
data['EngagementScore'] = (
    data['SocialShares'] * 1.5 +
    data['WebsiteVisits'] +
    data['PagesPerVisit'] * 2 +
    data['TimeOnSite'] * 0.5)
```

```python
bins = [0, 25, 35, 45, 60, 100]
labels = ['18–25', '26–35', '36–45', '46–60', '60+']
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='CampaignChannel', y='ClickThroughRate', data=data, palette='Blues_d')
plt.title('Click-Through Rate by Campaign Channel')
plt.xlabel('Campaign Channel')
plt.ylabel('Average Click-Through Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='CampaignType', y='ConversionRate', data=data, palette='Greens_d')
plt.title('Conversion Rate by Campaign Type')
plt.xlabel('Campaign Type')
plt.ylabel('Average Conversion Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='AgeGroup', y='EngagementScore', data=data, palette='Purples_d')
plt.title('Engagement Score by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Engagement Score')
plt.tight_layout()
plt.show()
```

```python
data.to_csv('Customer Campaign Analysis', index=False)
```

PPC Campaigns have the Highest Click-Through Rates

Among all campaign channels, PPC (Pay-Per-Click) recorded the highest average click-through rate (~0.158%), slightly outperforming Social Media, Email, and SEO.

Insight: PPC is an effective channel for driving customer engagement at the ad level.

Awareness Campaigns Lead to Higher Conversion Rates

Campaigns designed for "Awareness" achieved the highest average conversion rate (~10.5%) compared to Conversion, Retention, and Consideration campaigns.

Insight: Building early brand awareness significantly improves downstream conversion.

Customers Aged 26–35 are the Most Engaged Group

The 26–35 age group showed the highest Engagement Score, followed by the 18–25 and 60+ segments.

Insight: Marketing efforts targeted at young adults (especially 26–35) could yield stronger engagement.

Email Engagement is Moderate but Effective

Email campaigns had a moderate click-through and open rate, suggesting that while email is a steady channel, it's slightly less effective than PPC or Social Media for immediate clicks.

Insight: Email still plays a supportive but important role in nurturing leads.

Higher Website Interaction = Higher Engagement

Features like WebsiteVisits, PagesPerVisit, and TimeOnSite were heavily weighted in calculating Engagement Scores, indicating that website interaction is a strong indicator of customer interest.

Insight: Improving website content and user experience can directly boost engagement rates.



The marketing analysis reveals that PPC campaigns and awareness-focused strategies are most effective for initial engagement and conversion. Young adult customers (especially aged 26–35) are the most responsive audience segment, highlighting a clear demographic to target. Email marketing remains useful for nurturing relationships, but optimizing website interactions should be a priority for driving deeper engagement. Future marketing strategies should allocate more budget to PPC and awareness campaigns, while enhancing website experiences to maximize campaign ROI.
